```python
# Referans: https://github.com/codebasics/py/blob/master/ML/1_linear_reg/1_linear_regression.ipynb
# Lineer Regresyon kullanarak metrekare değerine göre ev fiyatının tahmin edilmesi
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Verimizi excel dosyası içerisinden okuyoruz.
veri = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/evfiyatlari.xlsx') # Dosya yolunuzu ayarlamanız gerekmektedir.
veri
```

```python
# Verimizi daha iyi görebilmek adına görselleştiriyoruz.
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(veri.area,veri.price,color='red',marker='+')
```

```python
# Değerleri, lineer regresyon modelimize vermek için ayırıyoruz
metrekare = veri.drop('price',axis='columns')
metrekare
```

```python
fiyat = veri.price
fiyat
```

```python
# Linner regresyon formulü: y = mx + b
# Lineer regresyon modelimizi elimizdeki değerler ile çalıştırıyoruz. Bu şekilde formul içerisinde bulunan 'm' ve 'b' sabit değerlerini buluyoruz.
model = linear_model.LinearRegression()
model.fit(metrekare,fiyat)
```

```python
# Lineer regresyon formulü ile tahmin işlemini gerçekleştiriyoruz.
# Lineer regresyon formulünde x=3300 yerine koyularak işlem gerçekleşmektedir.
model.predict([[3300]])
```

```python
# Formul içindeki 'm' değeri 
model.coef_
```

```python
#Formul içindeki 'b' değeri 
model.intercept_
```

```python
# predict() metodu ile elde edilen sonucun sağlamasını gerçekleştirelim.
# y = mx + b
135.78767123 * 3300 + 180616.43835616432
```

```python
# Son olarak elimizde bulunan değerler ile oluşturulan lineer tahmin modelinin grafik üzerinde görsel olarak görüntüleyelim.
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(veri.area,veri.price,color='red',marker='+')
plt.plot(veri.area,model.predict(veri[['area']]), color = 'blue')

### Örneğimizin sonuna geldik :)
```
